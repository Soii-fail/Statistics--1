# Theory Questions

1) What is hypothesis testing in statistics?

-> Hypothesis testing is a statistical method used to determine whether there is enough evidence to support a claim about a population. It involves 1  formulating a null hypothesis (a statement of no effect) and an alternative hypothesis (a statement of an effect), and then using data to determine whether the null hypothesis can be rejected




2) What is the null hypothesis, and how does it differ from the alternative hypothesis?

-> The null hypothesis states there is no effect or no difference, while the alternative hypothesis states there is an effect or a difference.  We aim to disprove the null hypothesis in favor of the alternative.

3)  What is the significance level in hypothesis testing, and why is it important?

-> The significance level (alpha) is the probability of rejecting the null hypothesis when it's actually true (a Type I error).

It's important because it sets the threshold for how much evidence we need to reject the null hypothesis


4) What does a P-value represent in hypothesis testing?

->The p-value represents the probability of observing the data (or more extreme data) if the null hypothesis is true. A small p-value suggests strong evidence against the null hypothesis

5) How do you interpret the P-value in hypothesis testing?

-> The P-value in hypothesis testing measures the probability of obtaining results as extreme as the observed data, assuming the null hypothesis is true. A smaller P-value (typically ≤ 0.05) suggests strong evidence against the null hypothesis, leading to its rejection, while a larger P-value indicates insufficient evidence to reject it.

6)  What are Type 1 and Type 2 errors in hypothesis testing?

-> A **Type 1 error** (false positive) occurs when we reject a true null hypothesis, mistakenly concluding an effect exists. A **Type 2 error** (false negative) happens when we fail to reject a false null hypothesis, missing a real effect. Reducing one often increases the other, requiring a balance based on the problem context.

7) What is the difference between a one-tailed and a two-tailed test in hypothesis testing?

-> A **one-tailed test** checks for an effect in only one direction (e.g., greater than or less than), making it more sensitive to detecting directional differences. A **two-tailed test** checks for effects in both directions (e.g., significantly different, either higher or lower), requiring more evidence to reject the null hypothesis. The choice depends on whether we have a specific directional hypothesis.

8) What is the Z-test, and when is it used in hypothesis testing?

-> A **Z-test** is a statistical test used to determine whether there is a significant difference between sample and population means when the population variance is known and the sample size is large (**n ≥ 30**). It is commonly used for comparing means or proportions when data follows a normal distribution.

9) How do you calculate the Z-score, and what does it represent in hypothesis testing?

-> The Z-score is calculated as:
$$ Z = \frac{X - \mu}{\sigma} $$

where X is the sample value, μ is the population mean, and σ is the population standard deviation.

In hypothesis testing, the Z-score measures how many standard deviations a data point is from the population mean. A higher absolute Z-score indicates a more extreme deviation, helping determine whether to reject the null hypothesis based on the critical Z-values.

10) What is the T-distribution, and when should it be used instead of the normal distribution?

-> The **T-distribution** (Student’s t-distribution) is a probability distribution used when estimating population parameters with a **small sample size (n < 30)** and **unknown population standard deviation**. It is similar to the normal distribution but has heavier tails, allowing for more variability.  

We use the **T-distribution** instead of the normal distribution when:  
1. The sample size is **small**.  
2. The **population standard deviation (σ) is unknown**, so we estimate it using the sample standard deviation (s).  

As the sample size increases, the **T-distribution approaches the normal distribution**.

11)  What is the difference between a Z-test and a T-test?

| Feature        | Z-test | T-test |
|---------------|--------|--------|
| **When Used** | Large sample (**n ≥ 30**) | Small sample (**n < 30**) |
| **Population Standard Deviation (σ)** | Known | Unknown (estimated from sample) |
| **Distribution Used** | Normal distribution | T-distribution (heavier tails) |
| **Application** | Comparing means, proportions, large datasets | Comparing means when sample size is small |


12) What is the T-test, and how is it used in hypothesis testing?

-> The T-test is a statistical test used to compare the means of two groups and determine whether the difference is statistically significant. It is applied when the sample size is small (n < 30) and the population standard deviation is unknown.

In hypothesis testing, the T-test calculates a T-statistic and a P-value to determine if the null hypothesis (no difference between means) should be rejected. If P ≤ 0.05, we reject the null hypothesis, indicating a significant difference.

13) What is the relationship between Z-test and T-test in hypothesis testing?

The Z-test is used for large sample sizes (n ≥ 30) with a known population standard deviation, while the T-test is used for small sample sizes (n < 30) with an unknown population standard deviation. As sample size increases, the T-test approximates the Z-test.

14) What is a confidence interval, and how is it used to interpret statistical results?

-> A confidence interval (CI) is a range of values that likely contains the true population parameter, with a specified confidence level (e.g., 95%). It helps interpret statistical results by showing the uncertainty around an estimate; if the CI excludes the null hypothesis value, the result is considered statistically significant.




15) What is the margin of error, and how does it affect the confidence interval?

-> The **margin of error** is the range within which the true population parameter is likely to fall, calculated from the sample data. A larger margin of error increases the width of the confidence interval, reflecting more uncertainty, while a smaller margin of error narrows the interval, indicating more precision.

16)  How is Bayes' Theorem used in statistics, and what is its significance?

-> Bayes' Theorem is used to update the probability of a hypothesis based on new evidence. It combines prior knowledge (prior probability) with observed data (likelihood) to compute the posterior probability. Its significance lies in allowing us to revise our beliefs in light of new information, making it crucial in decision-making and predictive modeling.

17)What is the Chi-square distribution, and when is it used?

-> The **Chi-square distribution** is a probability distribution that represents the sum of squared standard normal variables. It is used in hypothesis testing, especially for **goodness-of-fit tests** and **tests for independence** in contingency tables, to assess whether observed frequencies differ significantly from expected frequencies.

18) What is the Chi-square goodness of fit test, and how is it applied?

-> The **Chi-square goodness of fit test** compares observed data with expected frequencies to determine if a sample comes from a specific population distribution. It is applied by calculating the Chi-square statistic and comparing it to a critical value or using the p-value to assess if the observed frequencies significantly differ from the expected ones.

19) What is the F-distribution, and when is it used in hypothesis testing?

-> The **F-distribution** is a probability distribution used to test hypotheses about variances. It is applied in **ANOVA** (Analysis of Variance) to compare the variances of two or more groups and determine if at least one group has a significantly different variance from the others. It is also used in regression analysis to test the overall significance of a model.

20) What is an ANOVA test, and what are its assumptions?

-> An **ANOVA test** (Analysis of Variance) compares the means of three or more groups to determine if at least one differs significantly. Its assumptions include:  
1. **Independence** of observations.  
2. **Normality** of the data in each group.  
3. **Homogeneity of variances** (equal variances across groups).


21) What are the different types of ANOVA tests?

-> The main types of **ANOVA tests** are:  

1. **One-way ANOVA** – Compares means of three or more groups based on one independent variable.  
2. **Two-way ANOVA** – Compares means based on two independent variables, also examining interaction effects.  
3. **Repeated measures ANOVA** – Used when the same subjects are measured multiple times under different conditions.

22) What is the F-test, and how does it relate to hypothesis testing?

-> The **F-test** is used to compare variances between two or more groups to determine if they are significantly different. In hypothesis testing, it helps assess the overall significance of a regression model (in ANOVA) or test if the variances are equal across groups, with a large F-statistic indicating a significant difference.

```python
# 1) Write a Python program to perform a Z-test for comparing a sample mean to a known population mean and interpret the results.

import numpy as np
from scipy import stats

# Sample data
sample_data = np.array([25, 30, 28, 22, 27, 29, 31, 24, 26, 30])
sample_mean = np.mean(sample_data)

# Known population mean and standard deviation
population_mean = 28
population_std = 3  # Example value (known)

# Z-test calculation
sample_size = len(sample_data)
z_score = (sample_mean - population_mean) / (population_std / np.sqrt(sample_size))

# Calculate p-value for two-tailed test
p_value = 2 * (1 - stats.norm.cdf(abs(z_score)))

# Results
print(f"Sample Mean: {sample_mean}")
print(f"Z-Score: {z_score}")
print(f"P-value: {p_value}")

# Interpretation
alpha = 0.05
if p_value <= alpha:
    print("Reject the null hypothesis: The sample mean is significantly different from the population mean.")
else:
    print("Fail to reject the null hypothesis: The sample mean is not significantly different from the population mean.")
```

```python
# 5) Create a Python function that calculates and visualizes Type 1 and Type 2 errors during hypothesis testing

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def visualize_errors(alpha=0.05, power=0.8, mu0=0, mu1=1, sigma=1):
    # Define critical value for Type 1 error
    critical_value = norm.ppf(1 - alpha)

    # Generate x values
    x = np.linspace(-3, 4, 1000)

    # Null and Alternative distributions
    y_null = norm.pdf(x, mu0, sigma)
    y_alt = norm.pdf(x, mu1, sigma)

    # Plot distributions
    plt.plot(x, y_null, label="Null (H0)", color='blue')
    plt.plot(x, y_alt, label="Alternative (H1)", color='green')

    # Shade Type 1 error region (α)
    plt.fill_between(x, y_null, where=(x >= critical_value), color='red', alpha=0.5, label="Type 1 Error (α)")

    # Shade Type 2 error region (β)
    beta_value = norm.cdf(critical_value, mu1, sigma)
    plt.fill_between(x, y_alt, where=(x <= critical_value), color='purple', alpha=0.5, label="Type 2 Error (β)")

    plt.axvline(critical_value, color='black', linestyle='dashed', label="Critical Value")
    plt.title("Type 1 (α) and Type 2 (β) Errors")
    plt.legend()
    plt.show()

    print(f"Type 1 Error (α): {alpha}")
    print(f"Type 2 Error (β): {beta_value}")

# Run function
visualize_errors()
```

```python
# 9) Write a Python function to calculate the confidence interval for a sample mean and explain its significance.

import numpy as np
import scipy.stats as st

def ci(data, c=0.95):
    """Confidence interval."""
    try:
        n = len(data)
        if n < 1: return None
        m = np.mean(data)
        se = st.sem(data)
        a = 1 - c
        df = n - 1
        cv = st.norm.ppf(1 - a/2) if n > 30 else st.t.ppf(1 - a/2, df)
        me = cv * se  # Margin of error
        return m - me, m + me
    except:
        return None

# Example
data = np.random.normal(10, 2, 50)
res = ci(data)
if res:
    l, u = res
    print(f"CI: ({l:.2f}, {u:.2f})")

data2 = []
res2 = ci(data2)
if res2:
    l2, u2 = res2
    print(f"CI: ({l2:.2f}, {u2:.2f})")

```

```python
# 10) Write a Python program to calculate the margin of error for a given confidence level using sample data.

import numpy as np
import scipy.stats as st

def margin_of_error(data, c=0.95):
    """Calculates the margin of error."""
    try:
        n = len(data)
        if n < 1: return None
        se = st.sem(data)
        a = 1 - c
        df = n - 1
        cv = st.norm.ppf(1 - a/2) if n > 30 else st.t.ppf(1 - a/2, df)
        return cv * se
    except:
        return None

# Example:
data = np.random.normal(10, 2, 50)
me = margin_of_error(data)
if me is not None:
    print(f"Margin of Error: {me:.2f}")

data2 = [] #example of empty data
me2 = margin_of_error(data2)
if me2 is not None:
    print(f"Margin of Error: {me2:.2f}")
```

```python
# 11) Implement a Bayesian inference method using Bayes' Theorem in Python and explain the process

import numpy as np

def bayes_theorem(prior, likelihood, evidence):
    """
    Calculates the posterior probability using Bayes' Theorem.

    Args:
        prior (float): The prior probability of the hypothesis.
        likelihood (float): The likelihood of the evidence given the hypothesis.
        evidence (float): The probability of the evidence.  (Sometimes called the marginal likelihood)

    Returns:
        float: The posterior probability of the hypothesis given the evidence.
               Returns None if any input is invalid (e.g., probabilities not in [0, 1]).
    """

    if not all(0 <= p <= 1 for p in [prior, likelihood, evidence]):
      return None # Check for valid probabilities

    if evidence == 0: # Avoid division by zero
        return None

    posterior = (prior * likelihood) / evidence
    return posterior


# Example usage:  Medical Diagnosis

# Hypothesis (H): The patient has a certain disease.
prior_disease = 0.01  # Prior probability of having the disease (1%)

# Evidence (E): The patient tests positive for the disease.
likelihood_positive = 0.95  # Probability of testing positive given the disease (95% sensitivity)
prob_positive = 0.05 # Probability of testing positive given *no* disease (5% false positive rate)
evidence_positive = (prior_disease * likelihood_positive) + ((1 - prior_disease) * prob_positive) # Probability of testing positive (marginal likelihood)

# Calculate the posterior probability (probability of having the disease given a positive test)
posterior_disease = bayes_theorem(prior_disease, likelihood_positive, evidence_positive)


if posterior_disease is not None:
  print(f"Posterior probability of disease given positive test: {posterior_disease:.4f}")

# Example 2: Spam filter
prior_spam = 0.2 # 20% of all emails are spam
likelihood_word = 0.8 # The word "free" appears in 80% of spam emails
prob_word = 0.1 # The word "free" appears in 10% of non-spam emails
evidence_word = (prior_spam * likelihood_word) + ((1- prior_spam) * prob_word)
posterior_spam = bayes_theorem(prior_spam, likelihood_word, evidence_word)

if posterior_spam is not None:
  print(f"Posterior Probability of spam given 'free' is in email: {posterior_spam:.4f}")

#Example 3: Invalid input
prior = 1.2
likelihood = 0.5
evidence = 0.3
posterior = bayes_theorem(prior, likelihood, evidence)
if posterior is not None:
  print(f"Posterior: {posterior}")
```

```python
# 12) Perform a Chi-square test for independence between two categorical variables in Python

import pandas as pd
from scipy.stats import chi2_contingency

# Example data
data = {'Variable1': ['A', 'B', 'A', 'A', 'B', 'B'],
        'Variable2': ['X', 'X', 'Y', 'Y', 'X', 'Y']}

df = pd.DataFrame(data)

# Contingency table
contingency_table = pd.crosstab(df['Variable1'], df['Variable2'])

# Chi-square test
chi2, p, dof, expected = chi2_contingency(contingency_table)

print(f"Chi2 Statistic: {chi2}")
print(f"P-value: {p}")
print(f"Degrees of Freedom: {dof}")
print(f"Expected Frequencies Table: \n{expected}")
```

```python
# 16) Implement an F-test using Python to compare the variances of two random samples

import numpy as np
from scipy.stats import f

# Generate two random samples
np.random.seed(42)
sample1 = np.random.normal(loc=50, scale=10, size=30)  # Mean=50, Std=10
sample2 = np.random.normal(loc=50, scale=15, size=30)  # Mean=50, Std=15

# Calculate variances
var1 = np.var(sample1, ddof=1)
var2 = np.var(sample2, ddof=1)

# Compute F-statistic
F_stat = var1 / var2 if var1 > var2 else var2 / var1  # Ensure F ≥ 1

# Compute p-value
df1, df2 = len(sample1) - 1, len(sample2) - 1
p_value = 1 - f.cdf(F_stat, df1, df2)

print(f"F-statistic: {F_stat}")
print(f"P-value: {p_value}")

# If p < 0.05, reject the null hypothesis (variances are significantly different).
```

```python
# 17) D Write a Python program to perform an ANOVA test to compare means between multiple groups and interpret the results

import numpy as np
from scipy.stats import f_oneway

# Generate sample data for three groups
np.random.seed(42)
group1 = np.random.normal(loc=50, scale=5, size=30)  # Mean=50
group2 = np.random.normal(loc=55, scale=5, size=30)  # Mean=55
group3 = np.random.normal(loc=60, scale=5, size=30)  # Mean=60

# Perform ANOVA test
F_stat, p_value = f_oneway(group1, group2, group3)

print(f"F-statistic: {F_stat}")
print(f"P-value: {p_value}")

# Interpretation
if p_value < 0.05:
    print("Significant difference between group means (reject null hypothesis).")
else:
    print("No significant difference between group means (fail to reject null hypothesis).")
```

```python
# 18)  Perform a one-way ANOVA test using Python to compare the means of different groups and plot the results

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import f_oneway

# Generate sample data for three groups
np.random.seed(42)
group1 = np.random.normal(loc=50, scale=5, size=30)  # Mean=50
group2 = np.random.normal(loc=55, scale=5, size=30)  # Mean=55
group3 = np.random.normal(loc=60, scale=5, size=30)  # Mean=60

# Perform ANOVA test
F_stat, p_value = f_oneway(group1, group2, group3)

print(f"F-statistic: {F_stat}")
print(f"P-value: {p_value}")

# Interpretation
if p_value < 0.05:
    print("Significant difference between group means (reject null hypothesis).")
else:
    print("No significant difference between group means (fail to reject null hypothesis).")

# Plot results
data = [group1, group2, group3]
labels = ["Group 1", "Group 2", "Group 3"]

plt.figure(figsize=(8, 5))
sns.boxplot(data=data)
plt.xticks(ticks=[0, 1, 2], labels=labels)
plt.title("Comparison of Group Means (ANOVA)")
plt.ylabel("Values")
plt.grid(True)
plt.show()
```

```python
# 19)  Write a Python function to check the assumptions (normality, independence, and equal variance) for ANOVA

import numpy as np
import scipy.stats as stats
from itertools import combinations

def check_anova_assumptions(*groups):
    # 1. Normality Test (Shapiro-Wilk test)
    print("Normality Test (Shapiro-Wilk):")
    for i, group in enumerate(groups, 1):
        stat, p = stats.shapiro(group)
        print(f"Group {i}: W={stat:.3f}, p={p:.3f} {'(Normal)' if p > 0.05 else '(Not Normal)'}")

    # 2. Independence Assumption (Assumed if data comes from independent sources)
    print("\nIndependence Assumption: Ensure groups are collected independently.")

    # 3. Homogeneity of Variances (Levene’s test)
    stat, p = stats.levene(*groups)
    print(f"\nEqual Variance Test (Levene’s): W={stat:.3f}, p={p:.3f} {'(Equal Variance)' if p > 0.05 else '(Not Equal Variance)'}")

# Generate sample data
np.random.seed(42)
group1 = np.random.normal(loc=50, scale=5, size=30)
group2 = np.random.normal(loc=55, scale=5, size=30)
group3 = np.random.normal(loc=60, scale=5, size=30)

# Check assumptions
check_anova_assumptions(group1, group2, group3)
```

```python
# 20) Perform a two-way ANOVA test using Python to study the interaction between two factors and visualize the results

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm
from statsmodels.formula.api import ols

# Generate sample data
np.random.seed(42)
data = pd.DataFrame({
    'Factor_A': np.repeat(['A1', 'A2'], 30),  # First factor (e.g., Treatment A vs. B)
    'Factor_B': np.tile(['B1', 'B2', 'B3'], 20),  # Second factor (e.g., Condition X, Y, Z)
    'Response': np.random.normal(loc=50, scale=5, size=60) +
                np.random.choice([0, 3, -3], size=60)  # Adding interaction effects
})

# Perform two-way ANOVA
model = ols('Response ~ C(Factor_A) + C(Factor_B) + C(Factor_A):C(Factor_B)', data=data).fit()
anova_table = sm.stats.anova_lm(model, typ=2)

# Print ANOVA results
print(anova_table)

# Visualization: Interaction plot
plt.figure(figsize=(8, 5))
sns.pointplot(x='Factor_A', y='Response', hue='Factor_B', data=data, capsize=0.1, dodge=True, markers=['o', 's', 'D'])
plt.title("Interaction Effect of Factor A and Factor B on Response")
plt.xlabel("Factor A")
plt.ylabel("Response")
plt.grid(True)
plt.show()
```

```python
# 21) Write a Python program to visualize the F-distribution and discuss its use in hypothesis testing

import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

# Define degrees of freedom
df1_values = [1, 5, 10]
df2 = 20
x = np.linspace(0, 5, 500)

# Plot F-distributions
plt.figure(figsize=(8, 5))
for df1 in df1_values:
    y = stats.f.pdf(x, df1, df2)
    plt.plot(x, y, label=f"df1={df1}, df2={df2}")

# Plot settings
plt.title("F-Distribution")
plt.xlabel("F-value")
plt.ylabel("Probability Density")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# 23)  Simulate random data from a normal distribution, then perform hypothesis testing to evaluate the means

import numpy as np
from scipy.stats import ttest_1samp

# Simulate random data from a normal distribution
np.random.seed(42)
data = np.random.normal(loc=50, scale=5, size=100)  # Mean=50, Std=5, n=100

# Perform one-sample t-test to compare the sample mean with a known population mean (e.g., 50)
population_mean = 50
t_stat, p_value = ttest_1samp(data, population_mean)

# Display results
print(f"T-statistic: {t_stat}")
print(f"P-value: {p_value}")

# Interpretation
if p_value < 0.05:
    print("Reject the null hypothesis: Sample mean is significantly different from the population mean.")
else:
    print("Fail to reject the null hypothesis: Sample mean is not significantly different from the population mean.")
```

```python
# 24)  Perform a hypothesis test for population variance using a Chi-square distribution and interpret the results

import numpy as np
from scipy.stats import chi2

# Simulate random data from a normal distribution
np.random.seed(42)
data = np.random.normal(loc=50, scale=5, size=100)  # Mean=50, Std=5, n=100

# Calculate sample variance
sample_variance = np.var(data, ddof=1)

# Population variance (assumed known for hypothesis testing)
population_variance = 25  # Example population variance (5^2)

# Perform Chi-square test for population variance
n = len(data)  # Sample size
chi_square_stat = (n - 1) * sample_variance / population_variance

# Set significance level
alpha = 0.05

# Calculate the critical value for Chi-square distribution
critical_value_lower = chi2.ppf(alpha / 2, df=n - 1)
critical_value_upper = chi2.ppf(1 - alpha / 2, df=n - 1)

# Display results
print(f"Chi-square statistic: {chi_square_stat}")
print(f"Critical values: ({critical_value_lower}, {critical_value_upper})")

# Interpretation
if chi_square_stat < critical_value_lower or chi_square_stat > critical_value_upper:
    print("Reject the null hypothesis: Population variance is significantly different from the assumed value.")
else:
    print("Fail to reject the null hypothesis: Population variance is not significantly different from the assumed value.")
```

```python
# 25) Write a Python script to perform a Z-test for comparing proportions between two datasets or groups

import numpy as np
from statsmodels.stats.proportion import proportions_ztest

# Simulate data for two groups (e.g., success counts in two groups)
# Group 1: 100 successes out of 200 trials
# Group 2: 120 successes out of 250 trials
successes = np.array([100, 120])  # Success counts
trials = np.array([200, 250])    # Number of trials

# Perform Z-test for proportions
z_stat, p_value = proportions_ztest(successes, trials)

# Display the results
print(f"Z-statistic: {z_stat}")
print(f"P-value: {p_value}")

# Interpretation
if p_value < 0.05:
    print("Reject the null hypothesis: Proportions are significantly different.")
else:
    print("Fail to reject the null hypothesis: Proportions are not significantly different.")
```

```python
# 26) Implement an F-test for comparing the variances of two datasets, then interpret and visualize the results

import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

# Simulate two datasets with different variances
np.random.seed(42)
data1 = np.random.normal(loc=50, scale=5, size=100)  # Variance=25 (5^2)
data2 = np.random.normal(loc=50, scale=10, size=100)  # Variance=100 (10^2)

# Perform F-test to compare variances
var1 = np.var(data1, ddof=1)  # Sample variance of data1
var2 = np.var(data2, ddof=1)  # Sample variance of data2
f_stat = var1 / var2  # F-statistic
df1 = len(data1) - 1  # Degrees of freedom for data1
df2 = len(data2) - 1  # Degrees of freedom for data2

# Calculate the p-value for the F-statistic
p_value = 1 - stats.f.cdf(f_stat, df1, df2)

# Display results
print(f"F-statistic: {f_stat}")
print(f"P-value: {p_value}")

# Interpretation
if p_value < 0.05:
    print("Reject the null hypothesis: The variances are significantly different.")
else:
    print("Fail to reject the null hypothesis: The variances are not significantly different.")

# Visualization of the distributions
plt.figure(figsize=(10, 6))

# Plot the first dataset
plt.hist(data1, bins=20, alpha=0.6, label='Dataset 1 (Variance=25)', color='blue')
# Plot the second dataset
plt.hist(data2, bins=20, alpha=0.6, label='Dataset 2 (Variance=100)', color='red')

# Labels and title
plt.title('Distribution of Two Datasets')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.legend(loc='upper right')

plt.show()
```

```python
# 27) Perform a Chi-square test for goodness of fit with simulated data and analyze the results.

import numpy as np
from scipy.stats import chisquare

# Simulate observed data (e.g., frequencies of categories)
observed = np.array([30, 10, 20, 40])  # Observed frequencies for each category

# Define the expected frequencies (based on the null hypothesis)
# Assuming an equal distribution among the categories
expected = np.array([25, 25, 25, 25])

# Perform the Chi-square goodness of fit test
chi_stat, p_value = chisquare(observed, expected)

# Display the results
print(f"Chi-square Statistic: {chi_stat}")
print(f"P-value: {p_value}")

# Interpretation
if p_value < 0.05:
    print("Reject the null hypothesis: The observed frequencies significantly differ from the expected frequencies.")
else:
    print("Fail to reject the null hypothesis: The observed frequencies do not significantly differ from the expected frequencies.")
```
